```python
import pandas as pd

# Crear un DataFrame de ejemplo
data = {'Comuna': ['Comuna1', 'Comuna2', 'Comuna1', 'Comuna3', 'Comuna2'],
        'NombreAlumno': ['Juan', 'MarÃ­a', 'Pedro', 'Ana', 'Carlos'],
        'Edad': [20, 21, 22, 20, 22],
        'Promedio': [7.5, 8.0, 6.8, 7.2, 8.5]}
df = pd.DataFrame(data)

# Usar groupby para agrupar por la columna "Comuna"
grupos_por_comuna = df.groupby('Comuna')

# Definir una funciÃ³n personalizada para seleccionar columnas de importancia
def seleccionar_columnas_de_importancia(grupo):
    return grupo[['NombreAlumno', 'Edad']]

# Aplicar la funciÃ³n personalizada a cada grupo
df_con_columnas_de_importancia = grupos_por_comuna.apply(seleccionar_columnas_de_importancia)

# Reiniciar el Ã­ndice para obtener un DataFrame plano
df_con_columnas_de_importancia.reset_index(drop=True, inplace=True)

print(df_con_columnas_de_importancia)
```
