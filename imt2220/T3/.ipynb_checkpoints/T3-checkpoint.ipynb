```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Cargar los datos desde el archivo CSV en la carpeta "data"
data = pd.read_csv('data/data_HW3.csv', header=None, skiprows=1, names=['X', 'y'])
print(data.columns)
# Extraer los valores de las columnas X y la variable objetivo y
X = data['X'].to_numpy()
y = data['y'].to_numpy()

# Convertir los valores de X a números (si es necesario)
X = X.astype(float)

# Inicializar los parámetros a y b de manera aleatoria
a = np.random.randn()
b = np.random.randn()

# Definir la función de pérdida L en términos de theta (a y b)
def loss_function(theta, X, y):
    a, b = theta
    loss = np.mean((a * X + b - y)**2)
    return loss

# Calcular el gradiente de la función de pérdida
def gradient(theta, X, y):
    a, b = theta
    gradient_a = np.mean((a * X + b - y) * X)
    gradient_b = np.mean(a * X + b - y)
    return np.array([gradient_a, gradient_b])

# Hiperparámetros del descenso de gradiente
learning_rate = 0.01
num_epochs = 100

# Descenso de gradiente estocástico
for epoch in range(num_epochs):
    for i in range(len(data)):
        random_index = np.random.randint(0, len(data))
        xi = X[random_index]
        yi = y[random_index]
        theta = np.array([a, b])
        gradient_theta = gradient(theta, xi, yi)
        theta -= learning_rate * gradient_theta
        a, b = theta

    current_loss = loss_function(theta, X, y)
    print(f"Epoch {epoch + 1}/{num_epochs}, Loss: {current_loss}")

print("Optimal values of a and b:", a, b)

# Gráfica de la recta ajustada
plt.scatter(X, y, c=y, cmap='viridis')
plt.colorbar()
x_range = np.linspace(X.min(), X.max(), 100)
y_pred = a * x_range + b
plt.plot(x_range, y_pred, color='red', linewidth=2)
plt.xlabel('X')
plt.ylabel('y')
plt.title('Recta ajustada a los datos')
plt.show()
```

```python
# SGD
def gradiente(theta, x, y):
    a, b, c = theta
    grad_a = 2 * (a * x**2 + b * x + c - y) * x**2
    grad_b = 2 * (a * x**2 + b * x + c - y) * x
    grad_c = 2 * (a * x**2 + b * x + c - y)
    return np.array([grad_a, grad_b, grad_c])
    

def SGD(x0):
    his_L=[]
    theta = np.zeros(3) # initial guess
    for k, xk in enumerate(x0):
        theta -= (1 /k if k >= 1 else 1) * gradiente(theta, xk[0], xk[1])
        his_L.append(np.sum(theta.copy()))
    return his_L, theta
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Cargar los datos desde el archivo CSV
data = np.genfromtxt('data/data_HW3.csv', delimiter=',', skip_header=1)
x = data[:, 0]  # Columna de características (x)
y = data[:, 1]  # Columna de valores objetivo (y)

# Definir la función de pérdida L
def loss(theta, x, y):
    m = len(y)
    a, b = theta
    predictions = a * x + b
    error = predictions - y
    loss = (1 / (2 * m)) * np.sum(error ** 2)
    return loss

# Algoritmo de descenso de gradiente estocástico (SGD)
def stochastic_gradient_descent(x, y, learning_rate, num_iterations):
    m = len(y)
    theta = np.random.rand(2)  # Inicializar theta aleatoriamente
    for _ in range(num_iterations):
        random_index = np.random.randint(m)
        xi = x[random_index]
        yi = y[random_index]
        gradient_a = (theta[0] * xi + theta[1] - yi) * xi
        gradient_b = theta[0] * xi + theta[1] - yi
        theta[0] -= learning_rate * gradient_a
        theta[1] -= learning_rate * gradient_b
    return theta

# Parámetros del algoritmo SGD
learning_rate = 0.01
num_iterations = 1000

# Ejecutar el algoritmo SGD
optimal_theta = stochastic_gradient_descent(x, y, learning_rate, num_iterations)

# Imprimir los resultados
print("Parámetros óptimos (a, b):", optimal_theta)
print("Valor de L(θ; X, y):", loss(optimal_theta, x, y))

# Graficar los datos y la recta ajustada
plt.scatter(x, y, label='Datos')
plt.plot(x, optimal_theta[0] * x + optimal_theta[1], color='red', label='Recta Ajustada')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
